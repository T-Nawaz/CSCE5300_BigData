The data used here is the titanic dataset previously used in the Active Learning 2.

## Methodology

The very first step is to load the dataset in to a variable named 'df' using the read_csv() function. We then use the info() function to get an idea of what kind of data we are dealing with and what columns are in the dataset. After that, we can manipulate it to get specific columns.

## Code Implementation 

```python
# Importing the library
import pandas as pd
```

```python
# Reading the csv file and saving it to a variable
df = pd.read_csv('data/titanic.csv')
```

```python
# Explore dataset information
df.info()
```

We can use square bracket notation to get specific columns. The format is like so, dataframe[index_1 : index_2]. In this format, we will get all data from index_1 until index_2. The index_2 data is not included so we get index_2 - 1 data but not index_2 data.

```python
# Getting the first 5 rows
df[0:5]
```

Using a negative sign in front of the index number lets us start the indexing from reverse order. -1 points to the very last item, -2 points to the second to last item. So, if we want the last three rows, we can do the following.

```python
# Getting the last 3 rows
df[-3:]
```

If we want to get a specific column data for these rows, we can do dataframe[index_1 : index_2]['column_name']

```python
df[0:3]['Age']
```

If we want to get a multiple column data for these rows, we can do dataframe[index_1 : index_2][['column_name_1','column_name_1']]. Notice that list of columns is being used when we need to display multiple columns

```python
df[0:3][['Age', 'Fare']]
```

Following the format of the previous example, if we pass the column name 'Age' as a list in this format - ['Age'], we will get the output in a different format compared to before

```python
df[0:3][['Age']]
```

If we want to select specific rows, we can use either loc or iloc. The difference between them is that loc is based on index label and iloc is based on absolute position.

```python
df.loc[[1,2,3], ['Age', 'Fare']]
```

```python
df.iloc[[1,2,3], [5,9]]
```

Another method is to use conditions to filter out rows. In the following example, only surviving passengers are listed

```python
df[df['Survived'] == 1]
```

## Output and Result Interpretation

There are a couple of interesting points to notice in the results. 

The output of df[0:3]['Age'] and df[0:3][['Age']] is different. The first output is a pandas Series and the second output is a pandas DataFrame.

The loc and iloc notation is similar in nature but different in execution. On needs position information to function, the other need label information.
